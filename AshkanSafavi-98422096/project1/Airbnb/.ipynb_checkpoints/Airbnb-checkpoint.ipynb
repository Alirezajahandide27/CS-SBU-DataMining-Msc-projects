```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
ds = pd.read_csv("AB_NYC_2019.csv")
ds.head()
```

```python
ds['neighbourhood_group'].value_counts().plot.pie(explode=[0,0,0,0,0],autopct='%1.1f%%',shadow=False)
```

```python
sns.catplot(x="neighbourhood_group",y='price' ,kind="bar", data=ds,ci=None)
```

```python
f , _map = plt.subplots(figsize=(16,8))
_map = sns.scatterplot(x=ds.latitude,y=ds.longitude,hue=ds.neighbourhood_group,palette='tab10')
```

```python
sns.catplot(x="room_type",y='price' , data=ds,)
```

```python
review = ds.sort_values('number_of_reviews',ascending=False)
top_v = review.loc[:,['neighbourhood','number_of_reviews']][:20]
top_v = top_v.groupby('neighbourhood').mean().sort_values('number_of_reviews',ascending=False).reset_index()
f , _map = plt.subplots(figsize=(12,6))
sns.catplot(x=top_v["neighbourhood"],y=top_v['number_of_reviews'] ,kind="bar", data=ds,ci=None , ax=_map)
```
