```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import datetime
```

```python
data_set = pd.read_csv('C:\\Users\\Naim\\Pictures\\results.csv')
data_set.head(20)
```

```python
data_set["Year"]=pd.to_datetime(data_set['date']).dt.year
data_set["Month"]=pd.to_datetime(data_set['date']).dt.month
data_set=data_set.drop(columns=['date'], axis=1)
data_set.head(20)
```

```python
data_set['total_score']=data_set['home_score']+data_set['away_score']
data_set.head(20)
```

```python
con=[(data_set['home_score']==data_set['away_score']),(data_set['home_score']>data_set['away_score']),(data_set['home_score']<data_set['away_score'])]
value=[0,1,2]
value2=['no_win',data_set['home_team'],data_set['away_team']]
data_set['result']=np.select(con,value)
data_set['who_win']=np.select(con,value2)
data_set.head(20)
```

```python
b_team=data_set.groupby('who_win').sum()
b_team=b_team[['home_score','away_score','total_score']].sort_values(by=['total_score'],ascending=False)
b_team=b_team.drop(['no_win'])
b_team.head(20)
```

```python
b_team[:20].plot.bar()
```

```python
b_team2=data_set.groupby('who_win').count()
b_team2=b_team2[['result']].sort_values(by=['result'],ascending=False)
b_team2=b_team2.drop(['no_win'])

b_team2.head(20)
```

```python
b_team2[:20].plot.bar()
```

```python
b_team3=pd.merge(b_team,b_team2, how='inner', on='who_win')
b_team3['teams']=b_team3.index
b_team3.head(20)
```

```python
match=data_set.groupby('home_team').count()+data_set.groupby('away_team').count()
match=match[['city']].sort_values(by=['city'],ascending=False)
match=match.rename(columns = {'city':'play_count'})
match['teams']=match.index
match.head(20)
```

```python
match[:20].plot.bar()
```

```python
b_team4=pd.merge(match,b_team3,how='inner', on='teams')
b_team4['win_per_game']=b_team4['result']/b_team4['play_count']
b_team4['goal_per_game']=b_team4['total_score']/b_team4['play_count']
b_team4=b_team4[['teams', 'play_count', 'total_score', 'result', 'win_per_game', 'goal_per_game']]
b_team4.head(20)
```

```python
b_team4=b_team4.sort_values(by=['win_per_game'],ascending=False)
b_team4.head(20)
```

```python
b_team5=b_team4[(b_team4['win_per_game']>0,4) and (b_team4['play_count']>800)]
b_team5.head(25)
```

```python
team=data_set[['Year','who_win']]
team.head(20)
```

```python
def find(who_win):
    if "no_win" not in who_win.lower():
        return True
    return False

team2=team[team['who_win'].apply(find)]

team2.head()
```

```python
list1=[]
list2=[]
list3=[]
b=team2['Year'].count()
for i in range(0,b):
    list1.append(team2['Year'].iloc[i])
    list2.append(team2['who_win'].iloc[i])
    result=str(team2['Year'].iloc[i])+team2['who_win'].iloc[i]
    list3.append(result)
    
list_label = ["Year","who_win","Year_who_win"] 
list_col = [list1,list2,list3]       
zipped = zip(list_label,list_col) 
data_dict = dict(zipped)  

team3 = pd.DataFrame(data_dict)
team3.head(25)
```

```python
team4=team3.groupby('Year_who_win').count()
team4=team4[['who_win']].sort_values(by=['who_win'], ascending=False)
team4.head(25)
```

```python
team4[:25].plot.bar()
```

```python
data_c = data_set.drop(data_set[data_set.country == data_set.home_team].index)
list1=[]
list2=[]
b=len(data_c.index)
for i in range(0,b):
    list1.append(data_c['country'].iloc[i])
    list2.append(1)

list_label = ["Country","Num"] 
list_col = [list1,list2]       
zipped = zip(list_label,list_col) 
data_dict = dict(zipped) 

data_c3 = pd.DataFrame(data_dict) 
data_c4=data_c3.groupby('Country').count()
data_c5=data_c4[['Num']]
data_c6=data_c5.sort_values(by=['Num'], ascending=False)
data_c6.head(20)
```

```python
data_c6[:20].plot.bar()
```
