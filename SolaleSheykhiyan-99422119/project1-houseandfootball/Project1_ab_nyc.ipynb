رسم نمودار

```python


labels = ['Entire home/apt', 'Private room', 'Shared room']

review0means = [273.868229, 111.762283, 88.754777]
review1to25means = [202.014815, 85.760465, 64.866044]
reviewup25means=[182.482620, 80.001466, 58.014706]

x = np.arange(len(labels)) #label locations


width = 0.23  

fig, ax = plt.subplots()
rects1 = ax.bar(x - width, review0means, width, label='review 0')
rects2 = ax.bar(x  , review1to25means, width, label='review 1 to 25')
rects3 = ax.bar(x + width, reviewup25means, width, label='review up 26 ')
ax.set_ylabel('price')
ax.set_title('price by room type and review')
ax.set_xticks(x)
ax.set_xticklabels(labels)
ax.legend()



fig.tight_layout()

plt.show()
```
